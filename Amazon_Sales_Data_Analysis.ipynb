# Variables Description-

Based on the column names provided in the dataset, here's a description of each variable:

Region: The geographical region where the sales transaction occurred.

Country: The country where the sales transaction took place.

Item Type: The type or category of the item sold.

Sales Channel: The channel through which the sale was made (e.g., online, retail).

Order Priority: The priority level assigned to the order (e.g., high, medium, low).

Order Date: The date when the order was placed.

Order ID: A unique identifier for each order.

Ship Date: The date when the order was shipped.

Units Sold: The number of units of the item sold in the transaction.

Unit Price: The price per unit of the item sold.

Unit Cost: The cost per unit of the item sold.

Total Revenue: The total revenue generated from the sales transaction (units sold * unit price).

Total Cost: The total cost incurred for the sales transaction (units sold * unit cost).

Total Profit: The total profit generated from the sales transaction (total revenue - total cost).

These variables provide comprehensive information about each sales transaction, including details about the product, pricing, timing, and financial outcomes. Analyzing these variables can help identify trends, patterns, and insights that can inform strategic decision-making and optimization of sales strategies in the e-commerce domain.

# Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

```python
# importing visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

# Chart 1

```python
# Group data by 'Order Date' and sum the 'Total Profit' for each date
profit_over_time = df.groupby(pd.Grouper(key='Order Date', freq='Y'))['Total Profit'].sum()

# Plotting the profit trend over time
plt.figure(figsize=(10, 6))
profit_over_time.plot(marker='o', linestyle='-')
plt.title('Profit Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Insight-

As seen in the chart the profit was least in the year 2011 and was most peaked in the year 2012 after that  the business saw declining profits likely due to increased competition, economic downturns affecting consumer spending, shifts in consumer preferences, rising costs, pricing strategy changes, operational challenges, regulatory changes, and strategic decisions that didn't deliver expected returns.

# Chart 2

```python
# Group data by 'Region' and sum the 'Total Profit' for each region
profit_by_region = df.groupby('Region')['Total Profit'].sum().sort_values(ascending=False)

# Plotting the profit distribution by region
plt.figure(figsize=(10, 6))
profit_by_region.plot(kind='bar', color='lightgreen')
plt.title('Profit Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Insight-



*   Africa ranks first in profit distribution, indicating significant profitability possibly due to strong demand, favorable market conditions, or effective business strategies tailored to the region.
*   Europe follows Africa, suggesting a robust presence and profitability in European markets, potentially driven by established customer bases, effective distribution channels, or successful marketing efforts.


*   The Asia, Middle East, and North Africa region ranks third, showcasing diversified profitability across various continents and indicating the business's global reach and success in emerging markets.



*   Australia's mid-ranking suggests moderate profitability, highlighting potential growth opportunities or challenges specific to the Australian market.
*   The Caribbean ranks second to last, indicating relatively lower profitability, possibly due to factors such as a smaller market size, logistical challenges, or lower purchasing power.


*   North America's lower ranking may be surprising given its typically strong economy, suggesting specific challenges or market dynamics impacting the business in this region and signaling potential areas for improvement or growth.













# Chart 3

```python
# Grouping the data by country and calculating the total profit for each country
total_profit_by_country = df.groupby('Country')['Total Profit'].sum().sort_values(ascending=False)

# Plotting the total profit by country using a bar plot
plt.figure(figsize=(10, 6))
total_profit_by_country.plot(kind='bar', color='green')
plt.title('Total Profit by Country')
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.xticks(rotation=90, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Insight-
Visualizing total profit by country provides valuable insights into the geographic distribution of profitability within the business. By examining the bar plot, you can quickly identify which countries contribute the most to total profit and prioritize resources or strategies to further capitalize on these lucrative markets.

# Chart 4

```python
# Plotting a histogram of profit margins
plt.figure(figsize=(10, 6))
plt.hist(df['Profit Margin'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Profit Margins')
plt.xlabel('Profit Margin (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Insight-
The histogram show the frequency distribution of Profit Margin(%). This visualization can help you identify the most common profit margin ranges.

# Chart 5

```python
# Group data by sales channel and calculate the total profit for each channel
profit_by_channel = df.groupby('Sales Channel')['Total Profit'].sum()

# Plotting profit by sales channel using a bar plot
plt.figure(figsize=(10, 6))
colors = ['lightgreen' if channel == 'Offline' else 'skyblue' for channel in profit_by_channel.index]
profit_by_channel.plot(kind='bar', color=colors)
plt.title('Profit by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Insight-
The fact that offline channels generate more profit than online channels suggests that there may be untapped opportunities or advantages in the offline sales approach that contribute to higher profitability. It could imply that offline sales channels benefit from factors such as in-person customer interactions, brand visibility in physical locations, or established partnerships with retailers.

# Chart 6

```python
# Grouping the data by 'Item Type' and calculating total profit for each item type
profit_by_item_type = df.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False)

# Plotting the profit by item type using a bar plot
plt.figure(figsize=(10, 6))
profit_by_item_type.plot(kind='bar', color='orange')
plt.title('Total Profit by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Insight-
This bar plot shows the total profit generated by each item type, allowing to identify which types of items contribute the most to the total profit. We can then focus on strategies to maximize profitability for the top-performing item types like Cosmetics & Household or explore opportunities to improve profitability for lower-performing item types like Fruits,Meat, Snacks, etc.

# Chart 7

```python
# Group data by order priority and calculate the total profit for each priority level
total_profit_by_priority = df.groupby('Order Priority')['Total Profit'].sum().sort_values(ascending=False)

# Define a color palette for the bars
colors = sns.color_palette("viridis", len(total_profit_by_priority))

# Plotting the total profit by order priority using a bar plot with different colors
plt.figure(figsize=(10, 6))
total_profit_by_priority.plot(kind='bar', color=colors)
plt.title('Total Profit by Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Total Profit')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Insight-
The bar plot illustrating total profit by order priority reveals distinct profitability patterns across different priority levels. High-priority orders ('H') stand out as the most profitable, underscoring the importance of prioritizing their fulfillment to capitalize on their substantial contribution to overall profitability. While low-priority orders ('L') exhibit moderate profitability, they still make a positive impact on total profit, suggesting the need for efficient handling despite lower priority. Medium ('M') and critical ('C') priority orders demonstrate comparable profitability, highlighting the importance of maintaining efficient handling of both priority levels.
