# Variables Description

age: Age of the individual in years.

sex: Gender of the individual (0 = female, 1 = male).

cp: Chest pain type (0-3, indicating different types of chest pain).

trestbps: Resting blood pressure (in mm Hg) on admission to the hospital.

chol: Serum cholesterol level (in mg/dl).

fbs: Fasting blood sugar > 120 mg/dl (1 = true, 0 = false).

restecg: Resting electrocardiographic results (0-2).

thalach: Maximum heart rate achieved.

exang: Exercise-induced angina (1 = yes, 0 = no).

oldpeak: ST depression induced by exercise relative to rest.

slope: Slope of the peak exercise ST segment (0-2).

ca: Number of major vessels (0-3) colored by fluoroscopy.

thal: Thalassemia (1 = normal, 2 = fixed defect, 3 = reversible defect).

target: Presence of heart disease (1 = disease, 0 = no disease).

# Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

```python
# importing visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the style
sns.set(style="whitegrid")
```

# Chart 1

```python
# 1. Distribution of target variable
colors = ["#3498db", "#e74c3c"]
plt.figure(figsize=(10, 6))
sns.countplot(x='target', data=df, palette=colors)
plt.title('Distribution of Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.show()
```

Insight- This chart shows the distribution of individuals with and without heart disease. It helps to understand the prevalence of heart disease in the dataset. So we can see that people without heat disease is around 140 and with heart disease is 160.

# Chart 2

```python
# 2. Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Insight- From this chart we can see that individuals in the 50-60 age group are often at a higher risk for heart disease due to various factors such as prolonged exposure to risk factors like age-related physiological changes, and the cumulative effects of other health conditions like hypertension and diabetes. This concentration around the 50-60 age range suggests that the dataset is representative of a critical demographic that is highly relevant for studying heart disease.

# Chart 3

```python
# 3. Gender distribution
# Calculate the gender distribution
gender_counts = df['sex'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=['Male', 'Female'], autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.show()
```

Insight- The gender distribution plot shows approximately 31.8% females (0) and over 68.2% males (1), indicating a significant gender imbalance with a higher representation of males. This aligns with existing medical research suggesting that males are generally more prone to heart diseases compared to females.

# Chart 4

```python
# 4. Heart disease by age
boxplot_color = "#ff7f0e"  # Orange color
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='age', data=df, color=boxplot_color)
plt.title('Heart Disease by Age')
plt.xlabel('Heart Disease')
plt.ylabel('Age')
plt.show()
```

Insight-The boxplot visualizing heart disease by age reveals significant age-related trends in the prevalence of heart disease. Specifically, individuals without heart disease (target = 0) predominantly fall within the 52-62 age range. In contrast, individuals with heart disease (target = 1) are generally younger, with ages ranging from 45 to 58.

# Chart 5

```python
# 5. Chest pain type distribution
bar_colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]  # Blue, Orange, Green, Red
plt.figure(figsize=(10, 6))
sns.countplot(x='cp', data=df, palette=bar_colors)
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

The chest pain type distribution plot reveals the frequency of different chest pain types experienced by individuals in the dataset. Here is the breakdown based on the provided chart:

Type 0 (Typical Angina): 140 instances
Type 1 (Atypical Angina): 50 instances
Type 2 (Non-Anginal Pain): 85 instances
Type 3 (Asymptomatic): 25 instances

The distribution plot indicates that the most common chest pain type among the individuals is Typical Angina (Type 0), with 140 occurrences. This suggests that a significant portion of the population experiences chest pain that is classic and often associated with heart disease. Following this, Non-Anginal Pain (Type 2) is the next most frequent, with 85 instances, indicating a considerable number of individuals have chest pain that is not related to angina.

Atypical Angina (Type 1) accounts for 50 cases, suggesting less common but still notable occurrences of chest pain that doesn't fit the typical presentation of angina. Lastly, Asymptomatic (Type 3) is the least frequent with only 25 instances, highlighting that a smaller subset of individuals may have heart disease without experiencing any chest pain symptoms.

# Chart 6

```python
# 6. Maximum heart rate achieved by target
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='thalach', data=df)
plt.title('Maximum Heart Rate Achieved by Target')
plt.xlabel('Heart Disease')
plt.ylabel('Maximum Heart Rate Achieved')
plt.show()
```

Insight- The box plot depicting maximum heart rates achieved by individuals with and without heart disease (target 0 and 1, respectively) reveals distinct patterns in heart rate distribution based on disease status. Individuals without heart disease (target 0) exhibit maximum heart rates typically ranging from 125 to 150 beats per minute (bpm), while those with heart disease (target 1) demonstrate higher maximum heart rates, falling within the range of approximately 150 to 170 bpm. These findings align with clinical expectations, as elevated heart rates are often associated with compromised cardiovascular function in individuals with heart disease. The observed differences in maximum heart rates underscore the diagnostic value of heart rate measurements in distinguishing between individuals with and without heart disease. Furthermore, these insights have implications for treatment strategies, highlighting the importance of tailored interventions aimed at optimizing cardiac function and managing symptoms associated with elevated heart rates in individuals with heart disease.

```python
# 8. Scatter plot of age vs maximum heart rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='thalach', hue='target', data=df)
plt.title('Age vs Maximum Heart Rate Achieved')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate Achieved')
plt.show()
```

# Chart 9

```python
# 9. Blood pressure distribution
boxplot_color = "yellow"
plt.figure(figsize=(10, 6))
sns.boxplot(x='trestbps', data=df, color=boxplot_color)
plt.title('Resting Blood Pressure Distribution')
plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.show()
```

Insight- The box plot reveals that the majority of resting blood pressure (trestbps) values fall within the range of approximately 120 to 140 mm Hg, as indicated by the central box. This range represents the interquartile range (IQR), containing the middle 50% of the data. The median blood pressure, represented by the central line within the box, likely falls close to the center of this range. Additionally, the whiskers of the box plot extend to 1.5 times the IQR above and below the upper and lower quartiles, respectively. This suggests that while most blood pressure values are concentrated within the 120-140 mm Hg range, there are some outliers present in the dataset, as indicated by data points beyond the whiskers. These outliers may represent unusual or extreme blood pressure readings that deviate significantly from the typical range. Overall, the box plot provides a succinct summary of the distribution of resting blood pressure, highlighting the central tendency, variability, and presence of outliers in the dataset.

# Chart 10

```python
# 10. Cholesterol levels distribution
histplot_color = "green"
plt.figure(figsize=(10, 6))
sns.histplot(df['chol'], kde=True, color=histplot_color)
plt.title('Cholesterol Levels Distribution')
plt.xlabel('Cholesterol (mg/dl)')
plt.ylabel('Count')
plt.show()
```

Insight- The histogram plot of cholesterol levels distribution illustrates that the most frequent range of cholesterol levels in the dataset falls between 200 and 300 mg/dl, as evidenced by the peak in the histogram. This concentration suggests that a significant portion of individuals in the dataset have cholesterol levels within this range.
Cholesterol levels are a critical indicator of cardiovascular health, and values between 200 and 300 mg/dl are often considered borderline high to high, depending on individual risk factors and guidelines. The prominence of this range in the histogram underscores the prevalence of potentially elevated cholesterol levels within the population represented by the dataset.
